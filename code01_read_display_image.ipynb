```python
# Importing the opencv module
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read The Image

imageMascot = cv2.imread("mascot21.jfif") 

imageMascot_rgb = cv2.cvtColor(imageMascot, cv2.COLOR_BGR2RGB)
plt.imshow(imageMascot_rgb, cmap = plt.cm.gray)
plt.title('RGB_MascotOympics21')

plt.figure()
plt.title('Gray_MascotOympics21')
imageMascot_gray = cv2.cvtColor(imageMascot, cv2.COLOR_BGR2GRAY)
plt.imshow(imageMascot_gray, cmap = plt.cm.gray)
```
