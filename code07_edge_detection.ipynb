```python
# Importing the opencv module
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read The Image

imageMascot = cv2.imread("mascot21.jfif") 
print(imageMascot.shape)
imageMascot_rgb = cv2.cvtColor(imageMascot, cv2.COLOR_BGR2RGB)
plt.imshow(imageMascot_rgb)
plt.title('RGBMascotOympics21')
plt.figure()
```

```python
imageMascot_gray = cv2.cvtColor(imageMascot, cv2.COLOR_BGR2GRAY)

plt.imshow(imageMascot_gray, cmap = plt.cm.gray)

plt.title('GrayMascotOympics21')
plt.figure()
```

```python


#to detect edge
# Canny Edge Detection

edges_img = cv2.Canny(imageMascot_gray, threshold1=100, threshold2=200) # Canny Edge Detection

# Display Canny Edge Detection Image
plt.imshow(edges_img, cmap = plt.cm.gray)

plt.title('EDGEGrayMascotOympics21')
plt.figure()

#to detect edge
# Sobel Edge Detection

Sobeledges_imgx = cv2.Sobel(imageMascot_gray, ddepth=cv2.CV_64F, dx=1, dy = 0, ksize=3) # sobel Edge Detection
plt.imshow(Sobeledges_imgx, cmap = plt.cm.gray)

plt.title('XSobelEDGEGrayMascot')
plt.figure()

Sobeledges_imgy = cv2.Sobel(imageMascot_gray, ddepth=cv2.CV_64F, dx=0, dy = 1, ksize=3) # sobel Edge Detection

plt.imshow(Sobeledges_imgy, cmap = plt.cm.gray)

plt.title('YSobelEDGEGrayMascot')
plt.figure()


sobelxy = cv2.Sobel(imageMascot_gray, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3) # Combined X

# Display Canny Edge Detection Image
plt.imshow(sobelxy, cmap = plt.cm.gray)

plt.title('XYSobelEDGEGrayMascot')
plt.figure()

```
