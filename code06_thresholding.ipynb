```python
# Importing the opencv module
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Image thresholding
imagePark = cv2.imread("park.jfif") 
imagePark_gray = cv2.cvtColor(imagePark, cv2.COLOR_BGR2GRAY)
imagePark_rgb = cv2.cvtColor(imagePark, cv2.COLOR_BGR2RGB)

plt.imshow(imagePark_gray, cmap = plt.cm.gray)
plt.title('Gray Scenery')
plt.figure()
```

```python
#Image thresholding
plt.figure()
histg = cv2.calcHist(imagePark_gray,[0],None,[255],[0,256]) 

plt.plot(histg)
plt.show()
```

```python


#cv2.threshold(src, thresholdValue, maxVal, thresholdingTechnique)
ret, thresh_image = cv2.threshold(imagePark_gray, 200, 255, cv2.THRESH_BINARY)
plt.imshow(thresh_image, cmap = plt.cm.gray)
plt.title('Binary Image - High Threshold')
plt.figure()


```

```python


#cv2.threshold(src, thresholdValue, maxVal, thresholdingTechnique)
ret, thresh_image = cv2.threshold(imagePark_gray, 50, 255, cv2.THRESH_BINARY)
plt.imshow(thresh_image, cmap = plt.cm.gray)
plt.title('Binary Image - Low Threshold')
plt.figure()
```
