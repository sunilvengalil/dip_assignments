```python
# Importing the opencv module
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

imageSep = cv2.imread("foggy.jpeg") 
 

imageSep_gray = cv2.cvtColor(imageSep, cv2.COLOR_BGR2GRAY)
imageSep_rgb = cv2.cvtColor(imageSep, cv2.COLOR_BGR2RGB)

plt.imshow(imageSep_rgb, cmap = plt.cm.gray)
plt.title('Original Foggy')
```

```python


#Histogram Equalization
equ_sepia = cv2.equalizeHist(imageSep_gray)
plt.imshow(equ_sepia, cmap = plt.cm.gray)
plt.title('Equalized Foggy')
plt.figure()

```
