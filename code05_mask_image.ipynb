```python
# Importing the opencv module
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

imageSep = cv2.imread("sepia_gray.png") 
 

imageSep_gray = cv2.cvtColor(imageSep, cv2.COLOR_BGR2GRAY)


plt.imshow(imageSep_gray, cmap = plt.cm.gray)
plt.title('Sepia Gray')

```

```python

#Masking the image
hori_mask = np.zeros(imageSep_gray.shape)

[h, w] = imageSep_gray.shape

mask_height = int(np.floor(h/2))
hori_mask[0:mask_height,:]=1

vert_mask = np.zeros(imageSep_gray.shape)
mask_width = int(np.floor(w/2))
vert_mask[:,0:mask_width]=1

```

```python
plt.imshow(hori_mask, cmap = plt.cm.gray)
plt.title('HMask')
plt.figure()
```

```python
#New_sepia = cv2.bitwise_and(imageSep_gray, imageSep_gray, mask=mask_img)
New_sepia = imageSep_gray * hori_mask
plt.imshow(New_sepia, cmap = plt.cm.gray)
plt.title('Result_HMask')
plt.figure()
```

```python
plt.imshow(vert_mask, cmap = plt.cm.gray)
plt.title('VMask')
plt.figure()
```

```python
#New_sepia = cv2.bitwise_and(imageSep_gray, imageSep_gray, mask=mask_img)
New_sepia = imageSep_gray * vert_mask
plt.imshow(New_sepia, cmap = plt.cm.gray)
plt.title('Result_VMask')
plt.figure()
```
