```python
# Importing the opencv module
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imageRoses = cv2.imread("roses.jfif") 
#cv2.imshow("Roses", imageRoses) 
imageRoses_gray = cv2.cvtColor(imageRoses, cv2.COLOR_BGR2GRAY)
imageRoses_rgb = cv2.cvtColor(imageRoses, cv2.COLOR_BGR2RGB)
```

```python
type(imageRoses_gray)
```

```python
width, height, channels = imageRoses.shape
print("The width of the image is ", width)
print("The height of the image is ", height)
print("The number of channels are ", channels)
```

```python
plt.imshow(imageRoses_rgb, cmap = plt.cm.gray)
plt.title('Image Roses')
```

```python
#To extract the blue channel

imageRoses = cv2.imread("roses.jfif")
imageRosesBlue = imageRoses

imageRosesBlue[:,:,1]=0
imageRosesBlue[:,:,2]=0

imageRosesBlue_rgb = cv2.cvtColor(imageRosesBlue, cv2.COLOR_BGR2RGB)
imageRosesBlue_gray = cv2.cvtColor(imageRosesBlue, cv2.COLOR_BGR2GRAY)

plt.imshow(imageRosesBlue_rgb, cmap = plt.cm.gray)
plt.title('Image Roses Blue')


plt.figure()
plt.imshow(imageRosesBlue_gray, cmap = plt.cm.gray)
plt.title('Image Roses Blue Gray')
```

```python
imageRosesBlue_rgb.shape
```

```python

#To extract the red channel

imageRoses = cv2.imread("roses.jfif")

imageRosesRed = imageRoses

imageRosesRed[:,:,0]=0
imageRosesRed[:,:,1]=0

imageRosesRed_rgb = cv2.cvtColor(imageRosesRed, cv2.COLOR_BGR2RGB)
imageRosesRed_gray = cv2.cvtColor(imageRosesRed, cv2.COLOR_BGR2GRAY)


plt.imshow(imageRosesRed_rgb, cmap = plt.cm.gray)
plt.title('Image Roses Red')

plt.figure()
plt.imshow(imageRosesRed_gray, cmap = plt.cm.gray)
plt.title('Image Roses Red Gray')
```

```python
#To extract the green channel

imageRoses = cv2.imread("roses.jfif")

imageRosesGreen = imageRoses

imageRosesGreen[:,:,0]=0
imageRosesGreen[:,:,2]=0

imageRosesGreen_rgb = cv2.cvtColor(imageRosesGreen, cv2.COLOR_BGR2RGB)
imageRosesGreen_gray = cv2.cvtColor(imageRosesGreen, cv2.COLOR_BGR2GRAY)

plt.imshow(imageRosesGreen_rgb, cmap = plt.cm.gray)
plt.title('Image Roses Green')

plt.figure()
plt.imshow(imageRosesGreen_gray, cmap = plt.cm.gray)
plt.title('Image Roses Green Gray')
```
