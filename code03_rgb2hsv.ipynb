```python
# Importing the opencv module
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python


imageNG = cv2.imread("neeraj_gold.jfif") 

imageNG_rgb = cv2.cvtColor(imageNG, cv2.COLOR_BGR2RGB)

plt.imshow(imageNG_rgb, cmap = plt.cm.gray)
plt.title('Image NG RGB')
```

```python

hsvImage = cv2.cvtColor(imageNG, cv2.COLOR_BGR2HSV)


HueImg = hsvImage[:,:,0]
plt.imshow(HueImg, cmap="hsv")
plt.title('Image NG Hue')
plt.show()


SatImg = hsvImage[:,:,1]
plt.imshow(SatImg, cmap="Greys")
plt.title('Image NG Sat')
plt.show()


valueImg = hsvImage[:,:,2]
plt.imshow(valueImg, cmap="gray")
plt.title('Image NG Value')
plt.show()


```
